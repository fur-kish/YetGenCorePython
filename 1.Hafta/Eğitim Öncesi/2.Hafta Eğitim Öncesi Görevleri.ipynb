Kaç tane şekeri parçalayacaklarını hesaplayan bir Python kodu yazınız.

````python
def parcalanan_seker_sayisi(toplam_seker, kisi_sayisi):
    kisi_basina_dusen_seker = toplam_seker // kisi_sayisi
    kalan_seker = toplam_seker % kisi_sayisi
    return kalan_seker

toplam_seker = 91
kisi_sayisi = 3
parcalanan_seker = parcalanan_seker_sayisi(toplam_seker, kisi_sayisi)

print(f"Her bir kişi {kisi_sayisi} tane şeker alacak ve {parcalanan_seker} tane şeker parçalanacak.")
````
