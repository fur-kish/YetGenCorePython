# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def vki_hesapla(boy, kilo):
    boy_metrede = boy / 100  # Boyu metreye çevir
    vki = kilo / (boy_metrede ** 2)  # VKİ hesapla
    return vki

def vki_siniflandir(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "İleri derecede obez (morbid obez)"

def veri_girisi():
    isim = input("İsim: ")
    soyisim = input("Soyisim: ")
    yas = int(input("Yaş: "))
    boy = float(input("Boy (cm): "))
    kilo = float(input("Kilo (kg): "))

    vki = vki_hesapla(boy, kilo)
    sinif = vki_siniflandir(vki)

    return [isim, soyisim, yas, boy, kilo, vki, sinif]

def verileri_kaydet(dosya_adi, veri):
    with open(dosya_adi, mode='a') as file:
        file.write(",".join(map(str, veri)) + "\n")

def main():
    dosya_adi = "veriler.txt"
    
    with open(dosya_adi, mode='w') as file:
        file.write("İsim,Soyisim,Yaş,Boy (cm),Kilo (kg),VKİ,Sınıf\n")
    
    n = int(input("Kaç giriş yapacaksınız? "))
    
    for _ in range(n):
        veri = veri_girisi()
        verileri_kaydet(dosya_adi, veri)
    
    print(f"Veriler '{dosya_adi}' dosyasına kaydedildi.")

if __name__ == "__main__":
    main()
```

*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import csv
from datetime import datetime

print("YetBar otomatına hoş geldininz! Satın almak istediğiniz ürûnlerin numaralarını giriniz. ")

menu = [
    ["1", "Sade smoothie (Kendiniz tasarlayın)", 50],
    ["Smoothie için ek ürünler"],
    ["2", "Orman meyveleri", 40],
    ["3", "Hindistan cevizi", 15],
    ["4", "Mango", 35],
    ["5", "Bitter çikolata", 30],
    ["6", "Elma", 30],
    ["7", "Muz", 25],
    ["8", "Çilek", 35],
    ["9", "Ispanak", 25],
    ["10", "Karpuz", 35],
    ["11", "Fıstık ezmesi", 20],
    ["12", "Bal", 25]
]

def menu_yazdirma():
   menu_dosya_isim = "menu.cvs"
   with open(menu_dosya_isim, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Ürün Numarası", "Ürün", "Fiyat"]) 
    writer.writerows(menu)

def menu_gosterim():
  menu_dosya_isim = "menu.cvs"
  with open(menu_dosya_isim, mode='r', newline='') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)

def satin_alma():
    toplam_fiyat = 0
    secilen_urunler = []

    while True:
        urun_numarasi = input("Ürün numarasını girin (çıkmak için 'q'): ")
        if urun_numarasi.lower() == 'q':
            break
        for item in menu:
            if len(item) == 3 and item[0] == urun_numarasi:
                toplam_fiyat += item[2]
                secilen_urunler.append(item[1])
                print(f"Ürün eklendi: {item[1]}, Fiyat: {item[2]} TL")
    
    if secilen_urunler:
        print(f"\nToplam fiyat: {toplam_fiyat} TL")
        satin_alma_logla(secilen_urunler, toplam_fiyat)

def satin_alma_logla(urunler, fiyat):
    log_dosya_ismi = "satin_almalar.csv"
    with open(log_dosya_ismi, mode='a', newline='') as file:
        writer = csv.writer(file)
        tarih = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        writer.writerow([tarih, ', '.join(urunler), fiyat])
    print("Satın alma işlemi kaydedildi.")

def satin_almalar_goster():
    log_dosya_ismi = "satin_almalar.csv"
    with open(log_dosya_ismi, mode='r', newline='') as file:
        reader = csv.reader(file)
        for row in reader:
            print(row)

def main():
    while True:
        print("\nYapmak istediğiniz işlemi seçin:")
        print("1. Menüyü görüntüleme")
        print("2. Satın alma")
        print("3. Satın alma geçmişini görüntüleme")
        print("4. Çıkış")
        secim = input("Seçiminiz: ")

        if secim == '1':
            menu_gosterim()
        elif secim == '2':
            satin_alma()
        elif secim == '3':
            satin_almalar_goster()
        elif secim == '4':
            print("Çıkış yapılıyor...")
            break
        else:
            print("Geçersiz seçim, lütfen tekrar deneyin.")

if __name__ == "__main__":
    menu_yazdirma()
    main()
   





    
```
